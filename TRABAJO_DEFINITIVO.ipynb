```python
datc=[]
for j in range(0,len(LINK)):

  #i)
  r = requests.get(LINK[j])
  x = BeautifulSoup(r.text)
  #print(bs)
  x=str(x)

  #primera tabla:
  inicio="""<table border="2" bordercolor="#F5F5cC" cellpadding="3" cellspacing="0" width="100%">"""
  rini=len(inicio)
  termino="""</table>"""
  rter=len(termino)
  c=""
  c2=""

  for i in range(0,len(x)):
    if x[i:i+rini]==inicio:
      c=c+x[i:len(x)]
      #print(c)
      break
      
  for i in range(0,len(c)):
    if c[i:i+rter]==termino:
      c2=c2+c[0:i+8]
      #print(c2)
      break

  #segunda tabla:
  inicio2="""<table border="2" bordercolor="#feddfe" cellpadding="2" cellspacing="0" width="380">"""
  rini2=len(inicio2)
  termino2="""</table>"""
  rter2=len(termino2)
  c3=""
  c4=""
  
  for i in range(0,len(x)):
    if x[i:i+rini2]==inicio2:
      c3=c3+x[i:len(x)]
      #print(c3)
      break
      
  for i in range(0,len(c3)):
    if c3[i:i+rter2]==termino2:
      c4=c4+c3[0:i+8]
      #print(c4)
      break
      
  #tercera tabla:
  inicio3="""<table border="2" bordercolor="#E8e8Ff" cellpadding="2" cellspacing="0" width="380">"""
  rini3=len(inicio3)
  termino3="""</table>"""
  rter3=len(termino3)
  c5=""
  c6=""
  
  for i in range(0,len(x)):
    if x[i:i+rini3]==inicio3:
      c5=c5+x[i:len(x)]
      #print(c5)
      break
  
  for i in range(0,len(c5)):
    if c5[i:i+rter3]==termino3:
      c6=c6+c5[0:i+8]
      #print(c6)
      break

  trip=[]
  trip.append(c2)
  trip.append(c4)
  trip.append(c6)
  
  #print(trip)

  #datc=[]
  for i in trip:
    pri=[]
    bs=BeautifulSoup(i, "lxml")
    table_body=bs.find('table')
    rows = table_body.find_all('tr')
    for row in rows:
      cols=row.find_all('td')
      cols=[x.text.strip() for x in cols]
      #print(cols)
      pri.append(cols)
      #print(pri)
    datc.append(pri)

#print(datc)
#print(len(datc))





print(datc[167][2][1])





```
