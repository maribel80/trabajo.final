```python
import pandas as pd
from bs4 import BeautifulSoup as soup
import requests
from urllib.request import urlopen
url="https://www.ncbi.nlm.nih.gov/gene/?term=ABCG2%C2%A0(ATP-binding+cassette%2C+sub-family+G+(WHITE)%2C+member+2)"
us=urlopen(url)
paginahtml=us.read()
us.close()
print(paginahtml)
import pandas as pd
texto=[]
urls=[] 
df = pd.read_excel('1 Genes Produccin Leche por cromosoma Vacunos.xlsx') 
genasociado=df["Gen asociado"] 
for i in genasociado:
    i=i.replace(",","%2C").replace(" ","+").replace("\xa0","%C2%A0")
    texto.append(i)

for i in texto:
    urls.append("https://www.ncbi.nlm.nih.gov/gene/?term="+str(i))
    
urls.remove(urls[1])
urls
```

```python
for i in range(0,len(urls)):
    url=urls[i]
    
    print("\n"+genasociado[i]+"-------------") # se imprimir cada una de las filas del Gen asociado del excel
    
    us=urlopen(url)
    paginahtml=us.read()
    us.close()
    ssoup=soup(paginahtml, "html.parser")
    pagina=ssoup.find("div",{"class":"ui-ncbigrid-inner-div"})
    ubicacion=pagina.tbody.findAll("tr",{"class":"rprt"})
    for i in range(0,len(ubicacion)):
        if ubicacion[i].findAll("td")[2].text=="":
            print("na")
        else:    
            print(ubicacion[i].findAll("td")[2].text)
```
